```python
import pandas as pd
import numpy as np
import sklearn
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt


```

```python
path = r"C:\Users\apollo\Downloads\salary"
df = pd.read_csv(r"C:\Users\apollo\Downloads\salary\salary_predict_dataset.csv")
print(df)
```

```python
df['test_score'].fillna(df['test_score'].mean(), inplace = True)
df
```

```python
df['experience'].fillna(0, inplace = True)
df
```

```python
df['interview_score'].fillna(0, inplace = True)
df
```

```python
def strtono(word):
    dict= {0:0,'one':1,'two':2,'three':3,'four':4,'five':5,'six':6,'ten':10,'eleven':11,'twelve':12,'thirteen':13,'fifteen':15}
    return dict[word]
```

```python
df['experience'] = df['experience'].apply(lambda x: strtono(x))
df
```

```python
df.corr()
```

```python
x = df.iloc[:,:3]
x
```

```python
y = df['Salary']
y
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=45)
```

```python
from sklearn.linear_model import LinearRegression
mymodel = LinearRegression()
```

```python
mymodel.fit(x_train,y_train)
```

```python
y_pred = mymodel.predict(x_test)
```
