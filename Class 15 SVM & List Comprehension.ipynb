```python
import numpy as np
import pandas as pd
import sklearn
from sklearn.datasets import load_breast_cancer
```

```python
cancer = load_breast_cancer()
cancer.feature_names
```

```python
cancer.target_names
```

```python
df=pd.DataFrame(cancer.data, columns=cancer.feature_names)
df.head()
```

```python
df['target'] = cancer.target
df.head()
```

```python
df['cancer_class'] = df.target.apply(lambda x: cancer.target_names[x])
df
```

```python
import matplotlib.pyplot as plt
```

```python
y = df.target
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = df.drop(['target', 'cancer_class'],axis='columns')
Y = df.target
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size = 0.23, random_state = 6)
```

```python
from sklearn.svm import SVC
```

```python
mymodel = SVC()
```

```python
mymodel.fit(X_train, Y_train)
```

```python
mymodel.score(X_test, Y_test)
```

```python
from sklearn.metrics import confusion_matrix
```

```python
Y_pred = mymodel.predict(X_test)
```

```python
cm = np.array(confusion_matrix(Y_test,Y_pred))
```

```python
cm
```

```python
model_c = SVC(C=1)
model_c.fit(X_train,Y_train)
model_c.score(X_test,Y_test)
```

```python
model_kernel = SVC(kernel='linear')
model_kernel.fit(X_train,Y_train)
model_kernel.score(X_test,Y_test)
```
