# Exploratory Data Analysis


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv(r"C:\Users\apollo\Downloads\haberman.csv")
df
```

```python
df["status"].value_counts()
```

```python
df.isnull()
```

```python
sns.heatmap(df.isnull(),yticklabels=False)
```

```python
df.isnull().sum()
```

```python
sns.set_style('whitegrid')
sns.countplot(x='status',data=df)
```

```python
df.describe()
```

```python
#bivariate analysis
```

```python
sns.set_style('whitegrid')
sns.FacetGrid(df,hue='status').map(plt.scatter,"age","year").add_legend()
```

```python
sns.set_style("darkgrid")
sns.pairplot(df,hue="status",vars=["age","year","nodes"])
```

```python
#univariante
```

```python
live=df[df.status==1]
dead=df[df.status==2]
```

```python
plt.plot(live["age"],np.zeros_like(live["age"]),'o')
plt.plot(dead["age"],np.zeros_like(dead["age"]),'o')

plt.show()
```

```python
sns.FacetGrid(df,hue='status').map(plt.histplot,"year").add_legend()
```

```python
sns.distplot(df["age"])
```

```python
sns.histplot(df["age"])
```

```python
sns.boxplot(x='status',y='nodes',data=df)
```
